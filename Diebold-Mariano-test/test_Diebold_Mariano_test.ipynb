```python
import pandas as pd
from list_fun_DMtest import diebold_mariano

# ===== 1:XGBoost vs. HAR-D  (M(X), h=1, MSE, two-side) =====
xgb = pd.read_csv("XGB_X_1.csv")    # f_model     : XGBoost, M(X)
har = pd.read_csv("HARD_X_1.csv")   # f_benchmark : HAR-D,   M(X)

dm_stat, p_value, dbar = diebold_mariano(
    har["true_value"], xgb["prediction"], har["prediction"],
    loss="mse", h=1, two_sided=True)

print(f"  dm_stat = {dm_stat:.3f},  two-sided p = {p_value:.4g},  dbar = {dbar:.6f}")
# -> dm_stat=5.688, p=1.29e-08 (***); dbar>0
# => The two differ significantly; XGBoost is significantly more accurate than HAR-D.


# ===== 2:GRU M(X,T) vs. M(X)  (h=1, MSE, one-side) =====
mxt = pd.read_csv("GRUNet_RNNs_XaT_1.csv")  # f_model     : GRU, M(X,T) 
mx  = pd.read_csv("GRUNet_RNNs_X_1.csv")    # f_benchmark : GRU, M(X)   

dm_stat, p_value, dbar = diebold_mariano(
    mx["true_value"], mxt["prediction"], mx["prediction"],
    loss="mse", h=1, two_sided=False)

print(f"  dm_stat = {dm_stat:.3f},  one-sided p = {p_value:.4g},  dbar = {dbar:.6f}")
# -> dm_stat=-4.018, p≈1 (n.s.); dbar<0
# => Adding the textual feature T does not significantly improve the forecast.
```
